```python
import pandas as pd
import numpy as np

A = [[1,0.5,1],[2,1,2],[3,1.5,3],[4,2,4],[5,2.5,5],[6,3,6],[7,3.5,7],[8,4,8],[9,4.5,9],[10,5,10]]

num_rows = len(A)
distance_dict = {}
for i in range(num_rows):
    distance_dict[i] = {}
    for j in range(num_rows):
        distance_dict[i][j] = np.sqrt(np.sum(np.square(np.array(A[i]) - np.array(A[j]))))
distance_dict

```
